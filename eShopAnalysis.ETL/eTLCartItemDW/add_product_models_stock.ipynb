```python
"""
This cell to add all product models stock to the dotnet api
To add them to redis, since to add stock for product model, we have to add it in 
Web UI, and then the distributed transaction will be executed
But now since we crawl the data, we have to add them to redis directly through http request
"""
import requests
import os
import json
from dotenv import load_dotenv
from shared.get_connections import get_db_mongo
load_dotenv("config.env")

MONGODB_PRODUCT_COLLECTION = os.environ.get('MONGODB_PRODUCT_COLLECTION')
DOTNET_ADD_PRODUCT_MODELS_STOCK_ENDPOINT = os.environ.get('DOTNET_ADD_PRODUCT_MODELS_STOCK_ENDPOINT')
db = get_db_mongo()
product_collection = db.get_collection(MONGODB_PRODUCT_COLLECTION)
cursor_products = product_collection.find({})

post_request_payload = []
for p in cursor_products:
    product_models = p.get('ProductModels')
    product_model_id = product_models[0].get('_id')    
    product_id = p.get('_id')
    product_business_key = p.get('BusinessKey')
    stockToAdd = 100
    post_request_payload.append({
        "productId": product_id,
        "productModelId": product_model_id,
        "productBusinessKey": product_business_key,
        "currentQuantity": stockToAdd
    })
response = requests.post(DOTNET_ADD_PRODUCT_MODELS_STOCK_ENDPOINT, json=post_request_payload)
# for debugging
print("Request:", response.request.body)
print("Response content:", response.content)
print("Status code:", response.status_code)
print("Response:", response.reason)
```
